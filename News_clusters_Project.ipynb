# Fetch articles and headlines from news website

```python
import requests
from bs4 import BeautifulSoup
urllist = []
for i in range(1,11,1):
    url = "https://www.opindia.com/latest-news/page/" + str(i)
    urllist.append(url)
headlines = []
for j in urllist:
    data = requests.get(j)
    soup = BeautifulSoup(data.content,"html.parser")
    for i in soup.find_all("h3",class_="entry-title td-module-title"):
        headlines.append(i.find("a")["title"])
```

# Removal of Special Characters

```python
def rem_spl_chars(rev):
    import re
    rev = rev.lower()
    return re.sub("[^a-zA-Z ]","",rev)
```

# Word Tokenize for split the sentence into words

```python
def tokenize(A):
    from nltk.tokenize import word_tokenize
    A1 = word_tokenize(A)
    return A1
```

# Stop Word to remove all stop words from the corpus like : a, an, the etc..

```python
def stopword(A3):
    from nltk.corpus import stopwords
    sw = stopwords.words("english")
    A4 = []
    for i in A3:
        if i not in sw:
            A4.append(i)
    return A4
```

# Stemming Tchnique to remove suffix from every words

```python
def stem(A4):
    from nltk.stem import PorterStemmer
    ps = PorterStemmer()
    A5 = []
    for i in A4:
        A5.append(ps.stem(i))
    return A5
```

# Lemmatizer is used to get all words must match from its base word

```python

def lem(A5):
    from nltk.stem import WordNetLemmatizer
    wnlmt = WordNetLemmatizer()
    A6 = []
    for i in A5:
        A6.append(wnlmt.lemmatize(i))
    return A6
```

# Preprocessing of all the Words and Merge together

```python
def preproc(A):
    B = rem_spl_chars(A)
    B1 = tokenize(B)
    B2 = stopword(B1)
    B3 = stem(B2)
    B4 = lem(B3)
    return B4
```

# Data preparation and Fit a model
 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
X = tf.fit_transform(q)
q1 = tf.get_feature_names()
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
model = km.fit(X)
#B = pd.DataFrame(model.labels_,columns=["cluster"])
#B.cluster.value_counts()
```

```python
model.cluster_centers_.argsort()[:,:-40:-1]
```

# Find frequent keywords

```python
common_words = model.cluster_centers_.argsort()[:,:-40:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' :         ' + ','.join(q1[word] for word in centroid))
```
